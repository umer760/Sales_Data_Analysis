MERGING MONTH'S SALES INTO SINGLE FILE

```python

csv_files = glob.glob("C:\\Users\\Umer\\OneDrive\\Documents\\data_analysis.py\\project2.py\\Sales_data_2019"+"/*.csv") 
dataframe=[]
for file in csv_files:
    df=pd.read_csv(file)
    dataframe.append(df)
combine=pd.concat(dataframe,ignore_index=True)
combine.to_csv("Sales_Data.csv",index=True)
```

## DATA EXPLORATION AND CLEANING

```python
dataset=combine.copy()
dataset
```

```python
dataset=pd.DataFrame(dataset)
dataset.dropna(inplace=True)
dataset
```

```python
dataset=dataset.convert_dtypes()
```

```python
dataset.info()
```

```python
dataset["Month"]=dataset["Order Date"].str[0:2]
dataset
```

```python
dataset["Month"] = pd.to_numeric(dataset["Month"], errors='coerce')
r=dataset["Month"].isna().sum()
dataset.dropna(inplace=True)

```

```python
dataset["Month"].isna().sum()
```

```python
dataset["Price Each"]=dataset["Price Each"].astype(float)
```

```python
dataset["Quantity Ordered"]=dataset["Quantity Ordered"].astype(int)
```

```python
duplicates=dataset.duplicated()
dataset.drop_duplicates(inplace=True)
dataset
```

```python
dataset["Sales"]=dataset["Price Each"] * dataset["Quantity Ordered"]
dataset
```

```python
def monthname(x):
    if x==1:
     return "Janauary"
    elif x==2:
       return "Feburary"
    elif x==3:
       return "March"
    elif x==4:
       return "April"
    elif x==5:
       return "May"
    elif x==6:
       return "June"
    elif x==7:
       return "July"
    elif x==8:
       return "August"
    elif x==9:
       return "September"
    elif x==10:
       return "October"
    elif x==11:
       return "November"
    else:
       return "December"
```

## MONTHLY SALES

```python
dataset["Month_Name"]=dataset["Month"].apply(monthname)
```

```python
Month_sales=dataset.groupby("Month_Name")["Sales"].sum().reset_index(name="Month_Sales")
Month_sales=Month_sales.sort_values(by="Month_Sales",ascending=False)
Month_sales
```

```python
Month_Graph=Month_sales.plot.bar(x="Month_Name",y="Month_Sales",color="green")
def million(x,pos):
    return f"{x/1000000:.1f}M"
Month_Graph.yaxis.set_major_formatter(ticker.FuncFormatter(million))
for p in Month_Graph.patches:
    Month_Graph.annotate(f'{p.get_height()/1000:,.0f}K', (p.get_x() + p.get_width() / 2.,p.get_height()), ha='center', va='bottom')
plt.show()
```

## BEST TIME FOR SALE

###What time should we display advertisemens to maximize the likelihood of customerâ€™s buying product?

```python
dataset["Order_Time"]=dataset["Order Date"].str[9:]
dataset
```

```python
dataset["Order_Time"]=dataset["Order_Time"].astype(str)
dataset["Order Date"]=dataset["Order Date"].astype(str)
```

```python
def timezone(x):
    if  "00:00"<=x<="05:00":
        return "Mid_Night"
    elif "05:01"<=x<="12:00":
        return "Morning"
    elif "12:01"<=x<="15:00" :
        return "Afternoon" 
    else:
        return "Evening"
dataset["Time_Zone"]=dataset["Order_Time"].apply(timezone)
dataset            
                                        
```

```python
Best_selling_Time=dataset.groupby("Time_Zone")["Order ID"].nunique().reset_index(name="Customer")
Best_selling_Time
```

```python
selling_Time_Graph=Best_selling_Time.plot.bar(x="Time_Zone",y="Customer",color="green")
for p in selling_Time_Graph.patches:
    selling_Time_Graph.annotate(f'{p.get_height()/1000:,.0f}K', (p.get_x() + p.get_width() / 2.,p.get_height()), ha='center', va='bottom')
plt.show()
```

 BEST TIME FOR ADVERTISEMENT IS EVENING (3:01 PM TO 12:00PM)

## What products are most often sold together?

```python
dataset["Order ID"]=dataset["Order ID"].astype(int)
```

```python

Product=dataset.groupby("Order ID").agg({"Product":",".join})
Top_Ptoduct=Product.loc[Product["Product"].str.contains(",")].value_counts().reset_index(name="Best_Sold_Together")
print("Total rows: ",len(Top_Ptoduct))
Top_Ptoduct
```

Iphone and lightning are the two product which usually customer bought toghether.

```python
best_product=Top_Ptoduct.head(15)
TOP_Product_Graph=best_product.plot.barh(x="Product",y="Best_Sold_Together",color="green")
plt.bar_label(TOP_Product_Graph.containers[0])
plt.ylabel("TOP 15 Products")
plt.show()
```
